```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Heart_Disease_Prediction.csv')
over40 = data[data['Age'] > 40]
over40.head()
```

```python
colors = over40['Heart Disease'].map({'Presence': 1, 'Absence': 0})

fig, (plot1, plot2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# DESIGNING FIRST PLOT
figure = plot1.scatter(x=over40['Age'],
                       y=over40['Cholesterol'],
                       c=colors,
                       cmap='winter'
                     )
plot1.set(title="Age vs Cholesterol", xlabel='Age', ylabel='Cholesterol')
plot1.legend(*figure.legend_elements(), title="Disease (1)")
plot1.axhline(y=over40['Cholesterol'].mean(), linestyle='dotted')


# DESIGNING SECOND PLOT
figure = plot2.scatter(x=over40['BP'],
                       y=over40['Max HR'],
                       c=colors,
                       cmap='copper')
plot2.set(title='BP vs Max HR', xlabel='Blood Pressure', ylabel='Max Heart Rate')
plot2.legend(*figure.legend_elements(), title='Disease (1)')
plot2.axhline(y=over40['Max HR'].mean(), linestyle='dotted')


fig.suptitle(t="Patients Data Report", fontweight='bold', fontsize=32)


plt.style.use('seaborn-v0_8-darkgrid')
fig.tight_layout()
fig.savefig('Patients Data Report.png')
fig.show()
```
